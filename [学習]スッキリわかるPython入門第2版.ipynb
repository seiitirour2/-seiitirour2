```python
# リスト
scores = [10, 20, 30]
total = sum(scores)
average = total / len(scores)
print(f'{total = }\n{average = }\n{scores}')
print('-'*30)

# 追加
scores.append(40)
print(f'{total = }\n{average = }\n{scores}')
print('-'*30)

# 削除
scores.remove(30)
print(f'{total = }\n{average = }\n{scores}')
print('-'*30)

# スライスによる範囲指定（カンマの数を考えると分かりやすい）
a = [10, 20, 30, 40, 50, 60, 70, 80, 90]
print(a[1:3])
print(a[2:])
print(a[:3])
print(a[-3:-1])
```

```python
# ディクショナリ
# 同じキーは後の値へと上書きされる
# リストのような順番という概念は存在しないと考えると良い
scores = {'network':60, 'database':80, 'security':50, 'database':70}
print(scores)
print('-'*30)

# 追加と変更
scores['programing'] = 65
scores['security'] = 55
print(scores)
print('-'*30)

# 削除
del scores['security']
print(scores)
print('-'*30)

# キーや値のみを扱う
print(scores.keys())
print(scores.values())
print(sum(scores.values()))
```

```python
scores = [10, 20, 30, 40]
# while文
count = 0
while count < len(scores):
    print(f'検証する値： {scores[count]}')
    if scores[count] > 25:
        print('-合格')
    else:
        print('-不合格')
    count += 1
print('-'*30)

# for文
for data in scores:
    print(f'検証する値： {data}')
    if data >= 25:
        print('-合格')
    else:
        print('-不合格')
```

```python
ages = [20, 30, '無回答', 50, 60, 70, '無回答', 55, 45, 35, 25, 15]
num = 5
list = []
for data in ages:
    if not isinstance(data, int):
        continue
    if 20 <= data and data < 30:
        list.append(data)
        if len(list) == num:
            break
print(list)
```

```python
def eat(breakfast, lunch, dinner='カレー', *desserts):
    print(f'朝は{breakfast}を食べました')
    print(f'昼は{lunch}を食べました')
    print(f'夜は{dinner}を食べました')
    for d in desserts:
        print(f'おやつに{d}を食べました')

eat('トースト', 'パスタ', 'カレー', 'アイス', 'チョコ', 'カレー')
```

```python
def int_input(string):
    number = int(input(f'{string}を入力してください >>'))
    return number

def calc_payment(amount, people=2):
    dnum = amount / people
    pay = dnum // 100 * 100
    if dnum > pay:
        pay = int(pay + 100)
    return pay

def show_payment(pay, people, payorg):
    print('*** 支払額 ***')
    print(f'1人あたり{pay}円({people}人)、幹事は{payorg}円です')
    return None

people = int_input('人数')
amount = int_input('合計金額')

pay = calc_payment(amount, people)
payorg = amount - pay * (people - 1)

print(f'{pay = }, {people = }, {payorg = }')

show_payment(pay, people, payorg)
```
