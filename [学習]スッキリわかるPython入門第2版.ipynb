```python
scores = [10, 20, 30, 40]
# while
count = 0
while count < len(scores):
    print(f' {scores[count]}')
    if scores[count] > 25:
        print('-')
    else:
        print('-')
    count += 1
print('-'*30)

# for
for data in scores:
    print(f' {data}')
    if data >= 25:
        print('-')
    else:
        print('-')
```

```python
def eat(breakfast, lunch, dinner='', *desserts):
    print(f'{breakfast}')
    print(f'{lunch}')
    print(f'{dinner}')
    for d in desserts:
        print(f'{d}')

eat('', '', '', '', '', '')
```

```python
def int_input(string):
    number = int(input(f'{string} >>'))
    return number

def calc_payment(amount, people=2):
    dnum = amount / people
    pay = dnum // 100 * 100
    if dnum > pay:
        pay = int(pay + 100)
    return pay

def show_payment(pay, people, payorg):
    print('***  ***')
    print(f'1{pay}({people}){payorg}')
    return None

people = int_input('')
amount = int_input('')

pay = calc_payment(amount, people)
payorg = amount - pay * (people - 1)

print(f'{pay = }, {people = }, {payorg = }')

show_payment(pay, people, payorg)
```
